```python
import pandas as pd  
import numpy as np  
import sklearn as skl
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import random
```

```python
identity_df = pd.read_csv("fraud_dataset/train_identity.csv")
transaction_df = pd.read_csv('fraud_dataset/train_transaction.csv')
```

Merge the two datasets using TransactionID as the common feature

```python
full_df = pd.merge(transaction_df, identity_df, on="TransactionID", how="left")
```

```python
full_df.shape
```

Create two dataframes to seperate out is and is not Fraud

```python
fraud=full_df.loc[full_df['isFraud'] == 1]
non_fraud=full_df.loc[full_df['isFraud']==0]
```

```python
len(fraud)
```

```python
len(non_fraud)
```

```python
percent_fraud=100*(len(fraud)/len(full_df))
percent_fraud
```

Let's limit the number of features based off is more than 25% of the values are null.

```python
drop_df=pd.DataFrame({"na_count":full_df.isnull().sum().sort_values(ascending=False)})
```

```python
drop_df["per"]=drop_df["na_count"]/len(full_df)
```

```python
limit = .25
```

```python
drop_final=drop_df[drop_df["per"]<limit].index
full_df=full_df[full_df.columns.intersection(drop_final)]
```

```python
full_df.shape
```

Now we need to deal with non interger value. 

```python
full_df[full_df.select_dtypes(include=["object"]).columns].head(20)
```

```python
full_df = pd.get_dummies(full_df, columns = ['ProductCD', 'card4', 'card6', 'P_emaildomain'])
```

```python
full_df.shape
```

```python
full_df=full_df._get_numeric_data()
full_df=full_df.dropna()
```

```python
len(full_df[full_df['isFraud']==1])/len(full_df)*100
```

Let's split the dataset and get ready to train our linear regression model

```python
X_df=full_df.drop(["isFraud"], axis=1)
y_df=full_df['isFraud']
```

```python
X_train, X_test, y_train, y_test=train_test_split(X_df,y_df, test_size=.6)
```

```python
ax = fraud.plot.scatter(x='TransactionAmt', y='isFraud', color='Orange', label='Fraud')
non_fraud.plot.scatter(x='TransactionAmt', y='isFraud', color='Blue', label='Normal', ax=ax)
plt.show()
```

```python
x = full_df.iloc[:, :-1]
y = full_df['isFraud']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = .35)
```

```python
logistic=linear_model.LogisticRegression(C=1e5)
logistic.fit(X_train, y_train)
print('Score:', logistic.score(X_test, y_test))
```

```python
y_predicted = np.array(logistic.predict(X_test))
print(y_predicted)
```

```python
pd.crosstab(y_test, y_predicted, rownames=["True"], colnames=['Predicted'])
```
